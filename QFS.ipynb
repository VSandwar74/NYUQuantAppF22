Betting by a Percentage of Held Money Each Round (Pseudo-Kelly)

```python
# necessary imports
import math
import numpy as np
import random as ran
from math import trunc
import matplotlib.pyplot as plt

# define important variables
successCount = 0
totCount = 0
valueList = np.linspace(1, 0.01, num=100) # range of percents
successList = []

# main function
for value in valueList:
    print(value) # to track progress while running
    for i in range(10000): # 10,000 iterations of each value
        x = 100 # initial money
        while 1 == 1:
            # odds generation
            y = ran.randint(1, 10)
            if y <= 4:
                x += (value * x)
            elif y >= 5:
                x -= (value * x)

            x = trunc(x*100)/100 # truncate all values to cents

            # break conditions        
            if (value * x) < 0.01:
                totCount += 1 # always count one iteration of the loop
                break
            elif x > 199.99:
                successCount += 1 # count succesful 200$ returns
                totCount += 1
                break

    # calculate overall success rate for each value
    successRate = (successCount/totCount)
    print(successRate)
    # add to list for plotting
    successList.append(successRate)

# plot success rates against values
plt.plot(valueList, successList)
plt.title("Betting in Terms of Value (Pseudo-Kelly)")
plt.xlabel("Pecentage of Held Value Bet Each Turn")
plt.ylabel("Success Rate")
plt.show()
```

Betting using Martingale System

```python
# necessary imports
import math
import numpy as np
import random as ran
from math import trunc
import matplotlib.pyplot as plt

# important variables
successCount = 0
totCount = 0
valueList = np.linspace(1, 100.00, num=100) # varying initial bets
successList = []


# main loop
for value in valueList:
    # print(value)
    newValue = value # redefine term for each initial bet
    for i in range(10000):
        x = 100
        while 1 == 1:
            
            # odds generation
            y = ran.randint(1, 10)

            if y <= 4:
                x += (newValue)
                newValue = value # change value to inital if winning bet
            elif y >= 5:
                x -= (newValue)
                newValue *= 2 # double value on loss

            if x < value:
                totCount += 1
                break
            elif x > 199.99:
                successCount += 1
                totCount += 1
                break

    # calculate success for each value
    successRate = (successCount/totCount)
    successList.append(successRate) # list for plotting


# plot output
plt.plot(valueList, successList)
plt.title("Betting with Martingale System")
plt.xlabel("Initial Bet")
plt.ylabel("Success Rate")
plt.show()
```
